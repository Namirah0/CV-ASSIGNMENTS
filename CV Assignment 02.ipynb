```python
3. What does it mean to flatten? Is it necessary to include it in the MNIST CNN? What is the reason for this ?
Ans: Flattening is used to convert all the resultant 2-Dimensional arrays from pooled feature maps into a single long 
    continuous linear vector. The flattened matrix is fed as input to the fully connected layer to classify the image
    say you have a batch of N monochrome images which are R rows and C columns each. So your input matrix X has dimensions 
    NxRxC. And let’s say you want to do classification, and you have M classes, so you want an output matrix Y that is NxM.

Now multiply that by an ordinary 2-dimensional “dense” weights matrix in order to get a layer of H hidden activations. Your 
weights will need to have dimensions CxH, and your resulting activations will have dimensions NxRxH.

[NxRxC] mmult [CxH] = [NxRxH]

Notice the “R” is still there? In fact, for ordinary matrix multiplication (and barring CNNs), without doing *something* akin 
to flattening, you will never get rid of the “R”. The most you can hope for is an output matrix Y that is NxRxM.
```

```python
9. How does a color input interact with a convolution ?
Ans: Convolutional Neural Networks (CNNs) are a type of deep learning architecture that are particularly well-suited for image
    recognition tasks. In the case of color images, CNNs handle them by using multiple channels to represent the different 
    color channels (e.g. red, green, and blue) in the image. Each channel is processed separately by the CNN, and the results
    are combined to form a final output.The CNN starts by applying a set of filters (or kernels) to the input image. These
    filters are designed to detect specific features in the image, such as edges or textures. These features are then 
    represented by a set of feature maps, which are output by the CNN.The next step is a process called pooling, which reduces
    the size of the feature maps by selecting the most important features. This is done by taking a group of pixels and 
    replacing them with their maximum or average value.The final step is a process called fully connected layers, which takes
    the pooled feature maps and uses them to classify the objects in the image. The final output of the CNN is a set of 
    probabilities, one for each class (e.g. a dog, a cat, a car, etc.).Overall, CNNs are a powerful and widely-used technique
    for handling color images in object recognition tasks
```
